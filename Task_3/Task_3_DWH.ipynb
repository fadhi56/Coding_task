# Data Warehouse

The most commonly utlized schema for datawarehouse modeling are star and snowflake schema
Star schema required de-normaliation at the ETL layer and hence more time
whereas Snowflake schema increases dimension complexity and hence more time fro Analytics query.

For the current case, snowflake schema was selected to model the DWH. The choice is highly depndent on the information that we want to preserve. In our dataset there are a lot of movie ratings and user tags that change over the time. If we were not interested in this information, a star schema with 4 tables would have suffied. To capture more information, a snowflake schema was selected.

How to select the fact and dimension. Fact table usually contains the business facts, whereas the dimesnion table contains the description of this fact.
